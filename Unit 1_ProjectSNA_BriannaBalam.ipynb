```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
G = nx.read_edgelist('phonecalls.edgelist.txt', nodetype=int)

# Introduction
description = """
The dataset represents a network of phone calls between different nodes (individuals or entities). It is an undirected network where each edge represents a phone call between two nodes.
Understanding the structure of this network can reveal patterns in communication, identify key influencers, and help optimize network operations. Applications include telecommunications, marketing, and sociology.
"""
print(description)

# Visualization of the Network
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)
nx.draw_networkx(G, pos, with_labels=False, node_size=10, node_color='blue', edge_color='gray', alpha=0.6)
plt.title('Phone Call Network')
plt.show()
```

# Network characteristics

```python

# Calculate Network Characteristics
N = G.number_of_nodes()
E = G.number_of_edges()
density = nx.density(G)
avg_degree = sum(dict(G.degree()).values()) / N
clustering_coefficient = nx.average_clustering(G)

# Identify the largest connected component
largest_cc = max(nx.connected_components(G), key=len)
LCC = G.subgraph(largest_cc).copy()

# Calculate metrics for the largest connected component
avg_path_length = nx.average_shortest_path_length(LCC)
diameter = nx.diameter(LCC)

print(f"Number of Nodes: {N}")
print(f"Number of Edges: {E}")
print(f"Density: {density}")
print(f"Average Degree: {avg_degree}")
print(f"Clustering Coefficient: {clustering_coefficient}")
print(f"Average Path Length (LCC): {avg_path_length}")
print(f"Diameter (LCC): {diameter}")
```

```python
# Network Characteristics
size = G.number_of_nodes()
num_links = G.number_of_edges()
clustering_coeff = nx.average_clustering(G)

# Trabajar con la componente conexa más grande
largest_cc = max(nx.connected_components(G), key=len)
G_largest_cc = G.subgraph(largest_cc).copy()

# Calcular métricas en la componente conexa más grande
avg_path_length = nx.average_shortest_path_length(G_largest_cc)
diameter = nx.diameter(G_largest_cc)
eccentricity = nx.eccentricity(G_largest_cc)
radius = nx.radius(G_largest_cc)
periphery = nx.periphery(G_largest_cc)
center = nx.center(G_largest_cc)

print(f"Size of the Network: {size}")
print(f"Number of Links: {num_links}")
print(f"Average Path Length: {avg_path_length}")
print(f"Clustering Coefficient: {clustering_coeff}")
print(f"Diameter: {diameter}")
print(f"Eccentricity: {eccentricity}")
print(f"Radius: {radius}")
print(f"Periphery: {periphery}")
print(f"Center: {center}")

# Visualización de la componente conexa más grande
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G_largest_cc, seed=42, iterations=50)
nx.draw_networkx(G_largest_cc, pos, with_labels=False, node_size=10, node_color='blue', edge_color='gray', alpha=0.6)
plt.title('Largest Connected Component of the Phone Call Network')
plt.show()
```

# Centrality measures

```python
# Centrality Measures
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)

# Choosing Degree Centrality for this analysis
centrality = degree_centrality
sorted_centrality = sorted(centrality.items(), key=lambda x: x[1], reverse=True)
top_10_central_nodes = sorted_centrality[:10]

print("Top 10 Nodes by Degree Centrality:")
for node, centrality_value in top_10_central_nodes:
    print(f"Node {node}: {centrality_value}")
```

```python
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)
```

```python
import pandas as pd

# Convert centrality measures to DataFrame for visualization
centrality_df = pd.DataFrame({
    'Degree Centrality': degree_centrality,
    'Betweenness Centrality': betweenness_centrality,
    'Closeness Centrality': closeness_centrality,
    'Eigenvector Centrality': eigenvector_centrality
})

print(centrality_df.head())

# Plotting Centrality Measures
centrality_df.plot(kind='bar', figsize=(12, 8))
plt.title('Centrality Measures')
plt.xlabel('Nodes')
plt.ylabel('Centrality')
plt.show()
```

# Degree distribution

```python
# Degree Distribution
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_count = np.bincount(degree_sequence)
deg = np.arange(len(degree_count))

plt.figure(figsize=(10, 6))
plt.bar(deg, degree_count, width=0.80, color='b')
plt.title("Degree Distribution")
plt.ylabel("Count")
plt.xlabel("Degree")
plt.show()
```

```python
degrees = [G.degree(n) for n in G.nodes()]
degree_counts = np.bincount(degrees)
degrees = np.arange(len(degree_counts))

plt.figure(figsize=(10, 6))
plt.loglog(degrees, degree_counts, marker='o')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()
```

# Community detection

```python
from networkx.algorithms.community import greedy_modularity_communities

# Trabajar con la componente conexa más grande
largest_cc = max(nx.connected_components(G), key=len)
G_largest_cc = G.subgraph(largest_cc).copy()

# 6. Community Detection
communities = list(greedy_modularity_communities(G_largest_cc))
community_map = {}
for i, community in enumerate(communities):
    for node in community:
        community_map[node] = i

colors = [community_map[node] for node in G_largest_cc.nodes()]
pos = nx.spring_layout(G_largest_cc, seed=42, iterations=50)  # Calculando las posiciones para los nodos de la componente conexa más grande

plt.figure(figsize=(12, 12))
nx.draw_networkx(G_largest_cc, pos, node_color=colors, with_labels=False, node_size=10, cmap=plt.cm.rainbow)
plt.title("Community Detection in Largest Connected Component")
plt.show()
```
